```python
with open('bitcoinValues.csv','r') as f:
    data = []
    lines = f.readlines()
    for line in lines:
        line = line.replace('年','-')
        line = line.replace('月','-')
        line = line.replace('日','')
        data.append(line)
print(data)

with open('bitcoin_processed.csv','a') as f1:
    for line in data[::-1]:
        f1.write(line)

    
    
```
